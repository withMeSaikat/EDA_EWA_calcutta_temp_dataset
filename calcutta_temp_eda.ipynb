### Getting the data for the month January in 1995

```python
df_jan = df[df.year == 1995][df.month == 1].reset_index()
display(df_jan.head())
df_jan.plot(kind='scatter', x='day', y='temperature')
```

## Function to implement Exponentially Weighted Averages or Movements method on the new dataset of January temperatures

We can play around with the beta value which will fit the variation effectively

```python
def exp_avg_mvn(beta, df):
    arr = [0]
    for ind in range(1, df.shape[0]):
        temp = beta * df.iloc[ind - 1,:]['temperature'] + (1 - beta) * df.iloc[ind, :]['temperature']
        arr.append(temp)
    return arr
```

```python
pred_temp = exp_avg_mvn(0.245, df_jan)
```

```python
plt.figure(figsize=(15, 8))
predicted, = plt.plot(list(range(1, 31)),pred_temp[1:],linestyle = '--', label='Predicted Temperature Variation 1995')
actual, = plt.plot(df_jan['temperature'], label='Actual Temperature Variation 1995')
plt.legend(handles=[predicted, actual])
plt.xlabel('Days')
plt.ylabel('Temp(degree C)')
plt.title('Predicted vs Actual Temperature in Kolkata January 1995 using EWA')
plt.show()
```
